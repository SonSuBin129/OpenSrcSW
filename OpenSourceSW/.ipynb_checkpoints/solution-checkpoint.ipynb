```python
#1번

class Point:

    x=0

    y=0

    def __init__(self,a,b):

        self.x=a

        self.y=b

    

    def __add__(self,other):

        self.x=self.x+other.x

        self.y=self.y+other.y

        return Point(self.x,self.y)

    

    def show(self):

        print("({0:d},{1:d})".format(self.x,self.y))

    

    

    

a=Point(1,2)

a.show()



b=Point(3,4)

b.show()



a=a+b

a.show()

```

```python
#2번

import csv



f=open(r"C:\Users\손수빈\seoul.csv")

data=csv.reader(f)

next(data)



max_week=[] #일주일 단위로 일교차의 합들을 계산해서 넣을 배열

max_week_label=[] #일주일의 시작날짜와 마지막 날짜 넣을 배열

max_week_date="" #최종적으로 출력할 문자열



week=0 #7단위로 끊을 변수

diff_sum=0 #일교차 합을 구할 변수

start="" #일주일의 첫날

last="" #일주일의 마지막날



for row in data:

    if row[-1]!="":

        if row[-2]!="":

            row[-1]=float(row[-1])

            row[-2]=float(row[-2])

            if week<=7:

                diff_sum+=row[-1]-row[-2]

                week+=1

                if week==1:

                    start=row[0] #첫날

                if week==7:

                    last=row[0] #마지막날

                    max_week.append(diff_sum)

                    max_week_label.append(start+" ~ "+last)

                    week=0 #초기화

                    diff_sum=0 #초기화



max_diff=max(max_week) #일교차의 합이 가장 큰 값

index=max_week.index(max_diff) #일교차의 합이 가장 큰 값을 가지는 인덱스반환

max_week_date=max_week_label[index]

print("Max Week = "+max_week_date)

```

```python
#3번

import matplotlib.pyplot as plt

import numpy as np



def f_abs(array):

    result=0

    for i in array:

        result+=pow(i,2)

    result=np.sqrt(result)/len(array)

    return result



def Cos(doc, query):

    x=np.dot(doc,query.T) #분자가 될거임

    y=f_abs(doc)*f_abs(query)

    return x/y



query=np.random.randint(2,size=(1,6))



doc1=np.random.randint(2,size=(3,6)) #0과 1로만 이루어진 docs

doc2=np.random.randint(2,size=(3,6))

doc3=np.random.randint(2,size=(3,6))



r_doc1=Cos(doc1,query)

r_doc2=Cos(doc2,query)

r_doc3=Cos(doc3,query)



plt.rc("font",family="Malgun Gothic")

plt.title("Cosine similarities")

plt.bar(np.array([r_doc1,r_doc2,r_doc3]),np.array([r_doc1,r_doc2,r_doc3]))

plt.show()

    



 
```
