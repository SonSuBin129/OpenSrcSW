```python
#실습#1: 생일 기온 시각화

import csv
import matplotlib.pyplot as plt

f= open(r"C:\Users\손수빈\seoul2.csv")  #20110101~20201231에 해당하는 데이터가 seoul2.csv입니다.
data=csv.reader(f)
next(data)

highest=[]
lowest=[]

for row in data:
    if row[-1]!="" and row[-2]!="":
        if row[0].split('-')[1]=="01" and row[0].split('-')[2]=="29":
            highest.append(float(row[-1]))
            lowest.append(float(row[-2]))
            
plt.plot(highest,color="r",label="highest_temp")
plt.plot(lowest,color="b",label="lowest_temp")
plt.legend()
plt.show()
```

```python
# 실습#2: 최고 기온 시각화

import csv
import matplotlib.pyplot as plt

f= open(r"C:\Users\손수빈\seoul2.csv")  #20110101~20201231에 해당하는 데이터가 seoul2.csv입니다.
data=csv.reader(f)
next(data)

jan=[]
aug=[]

for row in data:
    month=row[0].split("-")[1]
    if row[-1]!="":
        if month=="01":
            jan.append(float(row[-1]))
        elif month=="08":
            aug.append(float(row[-1]))

plt.hist(aug, bins=100,color="r", label="Aug")
plt.hist(jan,bins=100, color="b", label="Jan")
plt.legend()
plt.show()
```
