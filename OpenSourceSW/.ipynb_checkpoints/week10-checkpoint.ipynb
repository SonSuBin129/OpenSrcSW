```python
# 실습#2: 최대값 구하기
str=input("Numbers? ")
list_a=str.split()
list_a=[int (i) for i in list_a]
check=max(list_a)
print("Max = {0:d}".format(check))
```

```python
# 실습#3: 암호화/복호화
def encryption(str):
    list_a=str.split()
    list_a.reverse()
    output=''
    for s in list_a:
        for c in s:
            output+=chr(ord(c)+1)
        output+=" "
    output.strip()
    print(output)
            
def decryption(str):
    list_a=str.split()
    list_n=[]
    
    for s in list_a:
        check=''
        for c in s:
            check+=chr(ord(c)-1)
        list_n.append(check)
    list_n.reverse()
    print(" ".join(list_n))
    
def menu():
    while(1):
        str= input("문자열을 입력하세요: ")
        choice=int(input("암호화(0), 복호화(1), 종료(2) 중에 선택하시오: "))
        if choice==0:
            encryption(str)
        elif choice==1:
            decryption(str)
        else:
            break
        
menu()    
```

```python
# 실습#4: 전화번호부
dic={}
for i in range(10):
    name=input("이름을 입력하시오: ")
    tel=int(input("전화번호를 입력하시오: "))
    dic[name]=tel

while(1):
    choice=int(input("검색 방법: 이름(0) 전화번호(1), 종료(2): "))
    if choice==0:
        check=input("이름을 입력하시오: ")
        if check in dic:
            print("{0:s}의 전화번호: {1:d}".format(check, dic.get(check)))
        else:
            print("해당하는 이름은 전화번호부에 없습니다.")
    elif choice==1:
        check=int(input("전화번호를 입력하시오: "))
        if check in dic.values():
            list=[]
            for i in dic.keys():
                if dic.get(i)==check:
                    list.append(i)
            print("{0:d}에 해당하는 이름: {1:s}".format(check," ".join(list)))
        else:
            print("해당하는 전화번호는 전화번호부에 없습니다.")
    else:
        print("종료합니다.")
        break
    
```
