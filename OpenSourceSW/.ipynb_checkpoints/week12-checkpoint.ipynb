```python
# 실습#1: 서울의 일교차

import csv
f= open(r"C:\Users\손수빈\seoul.csv")
data=csv.reader(f)
header=next(data)

max_vary=0
max_vary_d=""
max_temp=-999
max_temp_d=""
min_temp=999
min_temp_d=""

for row in data:
    if row[-1]=="":
        row[-1]=-999
    if row[-2]=="":
        row[-2]=999
    row[-1]=float(row[-1])
    row[-2]=float(row[-2])
    vary=row[-1]-row[-2]
    
    if max_temp<row[-1]:
        max_temp=row[-1]
        max_temp_d=row[0]
        
    if min_temp>row[-2]:
        min_temp=row[-2]
        min_temp_d=row[0]
        
    if max_vary<vary:
        max_vary=vary
        max_vary_d=row[0]
        
f.close()
print("=================일교차가 가장 큰 날=================")
print("날짜 : {0:s}  , 일교차 : {1:f}".format(max_vary_d,max_vary))
print("===============최저기온이 가장 낮은 날===============")
print("날짜 : {0:s}  , 최저기온 : {1:f}".format(min_temp_d,min_temp))
print("===============최고기온이 가장 높은 날===============")
print("날짜 : {0:s}  , 최고기온 : {1:f}".format(max_temp_d,max_temp))
```

```python
# 실습#2: 지하철 이용객

import csv
f= open(r"C:\Users\손수빈\subwayfee.csv",encoding="UTF8")
data=csv.reader(f)
header=next(data)
test_list=[]  #노선별로 구분하고 저장할 리스트
c=0  #리스트에 노선별로 구분하여 저장하기 위해 구분하는 변수
for row in data:
    if len(test_list)==0:
        c=1
    else:
        for row2 in test_list:
            if row[1]==row2[0][1]:
                row2.append(row)
                c=2
            else:
                c=1
    if c==1:
        a=[]
        a.append(row)
        test_list.append(a)
f.close()

for check in test_list:
    for row in check:
        row[4]=int(row[4].replace(",",""))
        row[5]=int(row[5].replace(",",""))
    check.sort(key=lambda x:(x[-1]+x[-2]))
    print("======= {0:s} ======".format(check[0][1]))
    print("<이용객이 가장 많은 역 순위>")
    for i in range(0,3):
        num=i+1
        print("{0:d}. {1:s} ({2:d}명)".format(num,check[-num][3],check[-num][4]+check[-num][5]))
    print("<이용객이 가장 적은 역 순위>")
    for i in range(0,3):
        print("{0:d}. {1:s} ({2:d}명)".format(i+1,check[i][3],check[i][4]+check[i][5]))
    print("\n")
    
```
