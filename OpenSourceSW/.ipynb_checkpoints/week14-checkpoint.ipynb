```python
#:  
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.float_format="{:.5f}".format

people=pd.read_csv(r"C:\Users\\people.csv", thousands=",")
people.set_index('', inplace=True)
people=people.iloc[1:,2:]  #  
people.columns=np.arange(0,101)
origin_p=people.astype(float)  #origin_p  

origin_p['']=people.std(axis=1)  #   

people=people-people.loc["  (1121500000)"].values.squeeze()
people=people**2
people[" "]=people.iloc[:,:].sum(axis=1) # 
people[""]=origin_p['']
origin_p.drop("",axis=1,inplace=True)  #origin_p  

people2=people[people[" "]>0]  #        


target=origin_p.loc["  (1121500000)"]
target_standard=people.at["  (1121500000)",""]
str_target=" ( : "+str(target_standard)+" )"

min_name=people2[" "].idxmin()
min_standard=people.at[min_name,""]
str_min=min_name+" ( : "+str(min_standard)+" )"
min=origin_p.loc[min_name] #      

max_name=people2[" "].idxmax()
max_standard=people.at[max_name,""]
str_max=max_name+" ( : "+str(max_standard)+" )"
max=origin_p.loc[max_name]  #      


################################# #######################################
plt.rc("font",family="Malgun Gothic")
plt.rcParams['figure.figsize'] = [18, 9]
plt.title("  ,     ")
plt.style.use("ggplot")
plt.plot(target, color="red", label=str_target)
plt.plot(min, color="blue", label=str_min)
plt.plot(max, color="gold", label=str_max)
plt.legend()
plt.show()
```
